```python
f = open("day10-21.txt", "r")
txt = f.readlines()
txt
```

```python
result = []
unfinished = []
for i in txt:
    expected = []
    found = False
    for j in i[:-1]:
            if j == "[":
                expected.append("]")
            elif j == "(":
                expected.append(")")
            elif j == "{":
                expected.append("}")
            elif j == "<":
                expected.append(">")
            else:
                res = expected.pop()
                if j != res:
                    if j == ")":
                        result.append(3)
                    elif j == "]":
                        result.append(57)
                    elif j == "}":
                        result.append(1197)
                    elif j == ">":
                        result.append(25137)
                    found = True
    if not found:
        temp = 0
        while expected:
            close = expected.pop()
            if close == "]":
                temp = temp*5 +2
            elif close == ")":
                temp = temp*5 +1
            elif close == "}":
                temp = temp*5 +3
            elif close == ">":
                temp = temp*5 +4
        unfinished.append(temp)
unfinished
```

```python
temp =0
for i in [3,57,1197,25137]:
    print(i)
    temp += i * result.count(i)
temp
```
