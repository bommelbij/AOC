```python
f = open("day4-21.txt", "r")
txt = f.readlines()
txt
```

```python
sequence = txt[0]
cards = txt[2:]
```

```python
bingo = []
card = []
for i in cards:
    if i != "\n":
        card.append(i[:-1].split())
    else:
        bingo.append(card)
        card = []
```

```python
def checkcard(card):
    for x in range(len(card)):
        temp = [row[0] for row in card]
        if all(isinstance(y, int) for y in temp):
            return True
        elif all(isinstance(y, int) for y in card[x]):
            return True
    return False
```

```python
sequence = sequence.split(",")
```

```python
def marknumber(card,number):
    for j in range(len(card)):
        for k in range(len(card[j])):
            if card[j][k] == number:
                card[j][k] = int(number)
    return card
```

```python

def callbingo():
    for i in sequence:
        count = 0
        for card in bingo:
            card = marknumber(card,i)
            if checkcard(card):
                return card,int(i)
            bingo[count] = card
            count += 1

res,call = callbingo()
   
```

```python
marked = 0
unmarked = 0 
for j in range(len(res)):
        for k in range(len(res[j])):
            if isinstance(res[j][k], int):
                marked += res[j][k]
            else:
                unmarked += int(res[j][k])
call*unmarked
```
