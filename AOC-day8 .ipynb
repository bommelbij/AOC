```python
def removevalue(diction,val):
    for key in diction.keys():
        temp = diction[key]
        if val in temp and not isinstance(temp, str):
            temp.remove(val)
        diction[key] = temp
    return diction
```

```python
def removeunique(direct):
    for key in direct.keys():
        if isinstance(direct[key],set) and len(direct[key]) == 1:
            temp = direct[key].pop()
            removevalue(direct,temp)
            direct[key] = str(temp)
    return direct
```
