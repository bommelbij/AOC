```python
short = []
for i in txt:
    row = []
    for j in i[:-1]:
        row.append(int(j))
    short.append(row)
```

```python
import math
paths = []
for a in range(len(short)):
    row = []
    for x in range(len(short[0])):
        row.append(math.inf)
    paths.append(row)
```

```python
paths[0][0] = 0
visited = [(0,0)]
```

```python
def dijkstra(values,distances,x,y,visited,nex):
    if x > 0 and (x-1,y) not in visited:
        if distances[x-1][y] == math.inf:
            distances[x-1][y] = distances[x][y]+values[x-1][y]
        elif distances[x-1][y] > distances[x][y]+values[x-1][y]:
            distances[x-1][y] = distances[x][y]+values[x-1][y]
        if (x-1,y) not in nex.keys():
            nex[(x-1,y)] = values[x-1][y]+ distances[x][y]
        elif nex[(x-1,y)] > values[x-1][y] + distances[x][y]:
            nex[(x-1,y)] = values[x-1][y] + distances[x][y]
    if x < len(values)-1  and (x+1,y) not in visited:
        if distances[x+1][y] == math.inf:
            distances[x+1][y] = distances[x][y]+values[x+1][y]
        elif distances[x+1][y] > distances[x][y]+values[x+1][y]:
            distances[x+1][y] = distances[x][y]+values[x+1][y]
        if (x+1,y) not in nex.keys():
            nex[(x+1,y)] = values[x+1][y]+ distances[x][y]
        elif nex[(x+1,y)] > values[x+1][y]+ distances[x][y]:
            nex[(x+1,y)] = values[x+1][y]+ distances[x][y]
    if y > 0 and (x,y-1) not in visited:
        if distances[x][y-1] == math.inf:
            distances[x][y-1] = distances[x][y]+values[x][y-1]
        elif distances[x][y-1] > distances[x][y]+values[x][y-1]:
            distances[x][y-1] = distances[x][y]+values[x][y-1]
        if (x,y-1) not in nex.keys():
            nex[(x,y-1)] = values[x][y-1]+ distances[x][y]
        elif nex[(x,y-1)] > values[x][y-1]+ distances[x][y]:
            nex[(x,y-1)] = values[x][y-1]+ distances[x][y]
    if y < len(values[0])-1  and (x,y+1) not in visited:
        if distances[x][y+1] == math.inf:
            distances[x][y+1] = distances[x][y]+values[x][y+1]
        elif distances[x][y+1] > distances[x][y]+values[x][y+1]:
            distances[x][y+1] = distances[x][y]+values[x][y+1]
        if (x,y+1) not in nex.keys():
            print(x,y+1)
            nex[(x,y+1)] = values[x][y+1]+ distances[x][y] 
        elif nex[(x,y+1)] > values[x][y+1] + distances[x][y]:
            nex[(x,y+1)] = values[x][y+1] + distances[x][y]
    #print(nex.keys())
    temp = ""
    mini = 100000
    for key in nex.keys():
        if nex[key] < mini:
            mini = nex[key]
            temp  = key
    del nex[temp]
    visited.append(temp)
    return distances,temp[0],temp[1],visited,nex
    
```

```python
x = 0 
y = 0
nex = {}
while x != len(short)-1 or y != len(short[0])-1:
    #print(x,y)
    #print(nex)
    paths,x,y,visited,nex = dijkstra(short,paths,x,y,visited,nex)
```
