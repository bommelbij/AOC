```python
res = 0
for i in txt:
    l,w,h = i.split("x")
    l=int(l)
    w= int(w)
    h = int(h)
    res += 2*l*w+2*w*h+2*l*h+min(l*w,h*w,h*l)
res
```

```python
f = open("day3-15.txt", "r")
txt = f.read()
```

```python
x = 0
y= 0
x1=0
y1=0
visited = [[0,0]]
for j,i in enumerate(txt):
    if j % 2 == 0:
        if i == "^":
            x +=1
        elif i == ">":
            y+= 1
        elif i == "v":
            x-=1
        else:
            y-= 1
        if [x,y] not in visited:
            visited.append([x,y])
    else:
        if i == "^":
            x1 +=1
        elif i == ">":
            y1 += 1
        elif i == "v":
            x1-=1
        else:
            y1 -= 1
        if [x1,y1] not in visited:
            visited.append([x1,y1])
    
len(visited)
```
